# Anomaly Detection & Time Series

# 1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
 - Anomaly Detection is the process of identifying data points, patterns, or behaviors that significantly deviate from what is considered normal in a dataset, making them suspicious or interesting for further investigation. It is widely used in fraud detection, cybersecurity, manufacturing defect detection, medical diagnosis, and system monitoring where unusual behavior can signal errors or threats. In anomaly detection, the system learns or defines what is “normal,” and then flags unusual data points that fall outside expected patterns. There are three major types of anomalies. **Point anomalies** refer to individual data points that are significantly different from the rest of the values. For example, if a bank customer suddenly withdraws ₹5,00,000 when their usual withdrawals are below ₹5,000, that transaction may be flagged as a point anomaly. **Contextual anomalies** are data points that appear normal in general but become anomalies when considering the context such as time, location, or season. For example, a temperature reading of 30°C may be normal in summer but anomalous in winter for a cold region, or electricity consumption that spikes at 3 AM when normally usage is low. These depend on context-based features. **Collective anomalies** occur when a group or sequence of related data points is abnormal even if individual points are normal by themselves. For example, a sequence of repeated failed login attempts on a server might indicate a cyberattack, or a sudden cluster of transactions from the same IP in a short time may indicate fraud. Thus, anomaly detection helps detect unexpected behavior by analyzing single values, context-dependent values, or unusual patterns across groups of data.


# 2.Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
 - Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are three popular unsupervised anomaly detection techniques that differ in how they identify outliers and in the scenarios where they perform best. **Isolation Forest** works by isolating data points through random partitioning using decision trees, assuming anomalies are easier to isolate due to being few and distant. It scales well for high-dimensional and large datasets, requires fewer parameters, and is suitable for fraud detection, network intrusion detection, and high-volume log analysis. **DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) detects anomalies by grouping points that lie in dense regions into clusters and marking low-density points as noise or outliers, making it powerful for datasets with arbitrary-shaped clusters and uneven distribution. Its performance depends heavily on parameters like eps and minPts, and it works well for spatial data, customer segmentation, and geographical anomaly detection. **Local Outlier Factor** identifies anomalies by comparing the local density of each point to that of its neighbors, meaning a point is considered anomalous if its density is significantly lower than that of surrounding points. LOF is useful when anomalies exist in local regions rather than global context, making it suitable for problems where density varies across clusters such as credit card fraud detection in region-specific distributions or fault detection in manufacturing sensor networks. In summary, Isolation Forest is best for large high-dimensional datasets, DBSCAN for density-based clustering and spatial anomalies, and LOF for detecting subtle anomalies depending on local neighborhood densities.


# 3.  What are the key components of a Time Series? Explain each with one example.
 - A time series consists of several key components that describe the structure and patterns hidden in sequential data collected over time, and understanding these helps in accurate forecasting and anomaly detection. The **trend component** represents the long-term upward or downward movement of data over time, such as steadily increasing sales of smartphones over several years due to rising demand and technology adoption. The **seasonal component** refers to regular periodic patterns that repeat at fixed intervals, for example, electricity consumption peaking in summer every year due to use of air conditioners or retail sales rising during festive seasons. The **cyclical component** captures fluctuations that occur over longer, irregular periods, usually influenced by economic or business cycles, such as recession-driven decline followed by recovery in automobile sales every few years, without a fixed length like seasonality. The **irregular or random component** represents unpredictable variations caused by unexpected events such as natural disasters, sudden supply chain disruptions, or pandemic lockdowns affecting production or sales for a brief period. Together, these components—trend, seasonality, cycles, and random noise—combine to explain how real-world time series behave and allow analysts to break down patterns for better forecasting and decision-making.


# 4. Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
 - In time series analysis, a series is said to be **stationary** when its statistical properties such as mean, variance, and autocorrelation remain constant over time, meaning the data does not exhibit long-term trends, seasonal patterns, or structural changes. Stationarity is crucial because many forecasting models like ARIMA assume that the underlying series is stationary for stable and reliable predictions. To test stationarity, analysts commonly use visual inspection through time plots, as well as statistical tests such as the Augmented Dickey–Fuller (ADF) test or the KPSS test: in the ADF test, a significant p-value indicates the series is stationary, whereas KPSS does the opposite. If the series is non-stationary, several transformations can make it stationary. **Differencing** subtracts previous observations from current ones to remove trends, such as converting a trending sales series into differences between consecutive months. **Log or power transformations** stabilize variance for datasets where fluctuations grow over time, such as stock prices. **Seasonal differencing** removes repeating seasonal patterns in monthly, quarterly, or yearly data. Smoothing methods like moving averages or detrending techniques can also remove deterministic trends. Once transformed, the resulting stationary series enables better modeling and more accurate time series forecasting.


# 5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
 - AR, MA, ARIMA, SARIMA, and SARIMAX are all time series forecasting models that differ mainly in how they incorporate past values, errors, seasonality, and external variables in their structure and applications. **Autoregressive (AR)** models rely only on linear combinations of past observations to predict future values, such as forecasting stock returns based on the previous few days' returns. **Moving Average (MA)** models instead use past forecast errors to capture short-term shocks, making them suitable when residuals exhibit autocorrelation. **ARIMA (Autoregressive Integrated Moving Average)** combines both AR and MA terms and adds differencing (the “I”) to remove trends, making it suitable for non-stationary series without seasonality, for example forecasting monthly sales after removing trend components. **SARIMA (Seasonal ARIMA)** extends ARIMA by incorporating seasonal AR, MA, and differencing terms to model repeating patterns in data such as quarterly GDP or yearly rainfall cycles, addressing both trend and seasonal effects. Finally, **SARIMAX (Seasonal ARIMA with exogenous variables)** further generalizes SARIMA by including external explanatory variables that influence the series, such as using advertising spend or diesel price to improve forecasts for retail demand or transportation costs. Thus, AR and MA are simpler linear models based on lagged values or errors, ARIMA handles non-stationarity, SARIMA adds seasonal structure, and SARIMAX enhances forecasting accuracy by integrating relevant external factors.


# 6. Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components
 - Time series decomposition is a method used to break down a time-dependent dataset into meaningful components to better understand underlying patterns and improve forecasting. When we load a dataset like the AirPassengers passenger count time series, the raw series contains several overlapping influences that make it difficult to analyze directly. Decomposition separates the series into three major components: the **trend** component, which captures the long-term movement such as the steady increase in the number of airline passengers over years; the **seasonality** component, which represents repeating periodic fluctuations caused by cyclic behavior such as holiday travel peaks occurring every year; and the **residual or irregular** component, which represents unpredictable random variations not explained by trend or seasonality, such as sudden drops or spikes due to weather, strikes, or economic shocks. By plotting and visually analyzing these separate components, analysts can identify structural patterns, detect anomalies, and build more accurate forecasting models such as ARIMA and SARIMA that rely on understanding underlying stationarity and seasonal behavior.


```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

# Load AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Create proper datetime index
date_index = pd.date_range(start="1949-01-01", periods=len(data), freq="M")

# Assign the index
data.index = date_index

# Extract passenger series
series = data['value']

# Plot original time series
plt.figure(figsize=(10,4))
plt.plot(series)
plt.title("AirPassengers – Original Time Series")
plt.ylabel("Passengers")
plt.xlabel("Year")
plt.show()

# Decompose the time series
result = seasonal_decompose(series, model='additive')

# Plot decomposition output
result.plot()
plt.show()
```

# 7.  Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
 - Isolation Forest is an unsupervised anomaly detection technique that works by isolating abnormal points instead of modeling normal data distribution, making it efficient for high-dimensional, large numerical datasets such as NYC Taxi fare transactions. The underlying intuition is that anomalies are rare and different from the majority of observations, so they require fewer random splits to isolate when recursively partitioning the data space using decision trees. During training, the model builds multiple random trees and computes anomaly scores based on the average path length: points that are isolated quickly receive higher anomaly scores and are flagged as outliers, while normal points require longer paths. Applying Isolation Forest to taxi fare data helps detect unusual trips, such as excessively high fares relative to trip distance or unexpected travel behavior. After fitting the model, two-dimensional scatter plots of distance versus fare amount visualize how anomalies fall outside dense normal clusters, providing intuitive insight into abnormal taxi trips that may indicate fraud, sensor error, or data entry mistakes.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# -------------------------------------------
# Example dataset (replace this with NYC data)
# -------------------------------------------

# sample 2-feature taxi-like numeric dataset
data = pd.DataFrame({
    "fare_amount":[5,10,7,9,6,8,11,50,60,55],   # anomaly fares
    "trip_distance":[1,2,1.5,2.2,1.3,1.8,2.1,10,12,11]  # anomaly distances
})

X = data[['fare_amount','trip_distance']]

# -------------------------------------------
# Train Isolation Forest
# -------------------------------------------
clf = IsolationForest(contamination=0.2, random_state=42)
clf.fit(X)

# Predict (-1 = anomaly, 1 = normal)
data['anomaly'] = clf.predict(X)

# -------------------------------------------
# Separate normal + anomaly points
# -------------------------------------------
normal = data[data['anomaly'] == 1]
anomalies = data[data['anomaly'] == -1]

# -------------------------------------------
# 2D scatter plot
# -------------------------------------------
plt.figure(figsize=(6,4))
plt.scatter(normal['fare_amount'], normal['trip_distance'], label="Normal points")
plt.scatter(anomalies['fare_amount'], anomalies['trip_distance'], label="Anomalies", marker='x')
plt.title("Isolation Forest – Anomaly Detection")
plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.legend()
plt.show()
```

# 8. Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
 - A SARIMA (Seasonal Autoregressive Integrated Moving Average) model is an extension of ARIMA that incorporates seasonal patterns in addition to trend and noise components, making it suitable for forecasting time series data that repeat at fixed intervals such as yearly fluctuations in air passenger numbers. The model is defined using non-seasonal terms (p,d,q) that capture autoregressive lags, differencing to remove trends, and moving-average lags, along with seasonal terms (P,D,Q,s) where “s” represents the seasonal period, such as 12 for monthly data with yearly seasonality. Training a SARIMA model on the monthly AirPassengers dataset allows the model to learn both long-term growth trends and strong seasonal travel patterns. Once trained, the fitted model can produce step-ahead forecasts, and visualizing the predicted next 12 months together with the historical time series makes it easier to observe whether the model captures recurring peaks in passenger counts and the continued upward trend. This forecasting approach is widely used in airline, retail, and transportation industries where seasonal patterns strongly influence demand.



```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.statespace.sarimax import SARIMAX

# ------------------------------
# Load AirPassengers Dataset
# ------------------------------
data = get_rdataset("AirPassengers").data

# Create monthly datetime index
data.index = pd.date_range(start="1949-01-01", periods=len(data), freq="M")

# Select passenger column
series = data['value']

# ------------------------------
# Fit SARIMA model
# (p,d,q) = (1,1,1)
# Seasonal order (P,D,Q,s) = (1,1,1,12)
# ------------------------------
model = SARIMAX(series,
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

results = model.fit()

# ------------------------------
# Forecast next 12 months
# ------------------------------
forecast = results.predict(start=len(series),
                           end=len(series)+11)

# ------------------------------
# Plot actual series + forecast
# ------------------------------
plt.figure(figsize=(10,5))
plt.plot(series, label="Historical Data")
plt.plot(forecast, label="12-Month Forecast", linestyle="--")
plt.title("SARIMA Forecast of AirPassengers")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```

# 9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
 - Local Outlier Factor (LOF) is an unsupervised anomaly detection technique that identifies unusual data points by measuring their local density compared to the density of their neighboring points. Instead of defining global boundaries for normal behavior, LOF focuses on the immediate neighborhood of each observation, calculating how isolated or sparse a point is relative to its neighbors. If a point lies in a region significantly less dense than surrounding points, it is assigned a higher anomaly score and labeled as an outlier. This makes LOF particularly useful for datasets where normal density varies across regions, and global distance-based methods might fail. When applied to numerical datasets such as transaction values or sensor readings, LOF can reveal subtle anomalies that cluster far from normal data. Visualizing detected anomalies using matplotlib helps analysts clearly see how LOF identifies isolated points away from dense clusters, supporting tasks like fraud detection, fault identification, or data quality assurance.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# ---------------------------------------
# Create small sample numeric dataset
# (can be replaced with real dataset)
# ---------------------------------------
data = pd.DataFrame({
    "x":[10, 12, 11, 13, 9, 12, 10, 200, 220, 250],   # anomalies
    "y":[20, 22, 19, 21, 18, 23, 20, 300, 280, 350]   # anomalies
})

X = data[['x','y']]

# ---------------------------------------
# Apply Local Outlier Factor
# ---------------------------------------
lof = LocalOutlierFactor(n_neighbors=3, contamination=0.2)

data['anomaly'] = lof.fit_predict(X)   # -1 = anomaly, 1 = normal

# Separate points
normal = data[data['anomaly']==1]
anomaly = data[data['anomaly']==-1]

# ---------------------------------------
# Scatter Plot visualization
# ---------------------------------------
plt.figure(figsize=(6,4))
plt.scatter(normal['x'], normal['y'], label="Normal")
plt.scatter(anomaly['x'], anomaly['y'], marker='x', label="Anomaly")
plt.title("Local Outlier Factor – Anomaly Detection")
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.legend()
plt.show()
```

# 10.  You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage
 - In a real-time power grid monitoring system, continuous 15-minute energy consumption streams from different regions must be analyzed to detect abnormal spikes or drops while simultaneously forecasting short-term demand. First, the streaming numerical features like timestamp and consumption can be preprocessed and smoothed, and an online anomaly detection method such as Isolation Forest or Local Outlier Factor (LOF) can identify unusual consumption points because these models can adapt to new incoming values by comparing their density or separation from historical patterns, whereas DBSCAN can cluster consumption values to mark low-density points as anomalies but requires batch windows. For short-term forecasting, SARIMAX is preferred because it captures seasonal patterns in daily and weekly power usage and can incorporate exogenous predictors like real-time weather, temperature, and regional events that strongly influence energy demand, while ARIMA/SARIMA treat only internal consumption patterns. Model validation in production requires rolling-window performance evaluation using metrics like MAE or RMSE, drift monitoring of residuals, retraining triggers when the error exceeds thresholds, and dashboard alerts when anomaly frequency surges or forecast confidence intervals widen. This combined solution provides operational value since accurate demand forecasts help load balancing, transformer scheduling, energy trading, and peak pricing strategies, while real-time anomaly detection prevents equipment failure, blackout risk, or cyber-attack compromise by responding early to unusual consumption patterns in the grid.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# --------------------------------------------------
# Fake streaming dataset (15-minute intervals)
# --------------------------------------------------
date_rng = pd.date_range(start='2024-01-01', freq='15min', periods=200)
usage = np.random.normal(200, 15, size=200)

# inject anomaly spikes
usage[50] = 350
usage[120] = 80

data = pd.DataFrame({"timestamp": date_rng, "energy_usage": usage})
data.set_index("timestamp", inplace=True)

# --------------------------------------------------
# 1) Real-time anomaly detection (Isolation Forest)
# --------------------------------------------------
model_iso = IsolationForest(contamination=0.03, random_state=42)
data["anomaly"] = model_iso.fit_predict(data[["energy_usage"]])

normal = data[data["anomaly"] == 1]
abnormal = data[data["anomaly"] == -1]

# Visualization
plt.figure(figsize=(10,4))
plt.plot(data.index, data["energy_usage"], label="Energy Usage")
plt.scatter(abnormal.index, abnormal["energy_usage"], color="red", label="Anomaly")
plt.title("Energy Usage with Anomaly Detection")
plt.legend()
plt.show()

# --------------------------------------------------
# 2) Short-term forecasting using SARIMAX
# --------------------------------------------------
train = data["energy_usage"]

sarimax = SARIMAX(train,
                  order=(1,1,1),
                  seasonal_order=(1,1,1,96)) # 96 points per day (96*15min)

results = sarimax.fit(disp=False)

forecast = results.predict(start=len(train), end=len(train)+12)

# Visualization
plt.figure(figsize=(10,4))
plt.plot(train[-200:], label="Historical")
plt.plot(forecast, label="Forecast (Next 12 intervals)", linestyle="--")
plt.title("SARIMAX Forecast for Real-Time Grid Load")
plt.legend()
plt.show()
```
